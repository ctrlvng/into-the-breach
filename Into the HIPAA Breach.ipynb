```python
# make a duplicate df, just in case

hack_df2 = hack_df
```

```python
# take a sample of 50 rows

hack_df3 = hack_df.sample(n=50)
hack_df3
```

```python
# create a .csv of the sample for manual categorization

hack_df3.to_csv('sample.csv')
```

```python
# categorize using SQL, to count multiple keywords at once

mysql = lambda q: sqldf (q, globals())
query1 = ('''
SELECT *, (CASE 
WHEN "Web Description" LIKE '%phishing%' THEN '1'
ELSE 0
END) AS 'Phishing',
(CASE 
WHEN "Web Description" LIKE '%ransomware%' THEN '1'
ELSE 0
END) AS 'Ransomware',
(CASE 
WHEN "Web Description" LIKE '%cyber%' THEN '1'
ELSE 0
END) AS 'Cyberattack',
(CASE 
WHEN "Web Description" LIKE '%unauthorized%' THEN '1'
ELSE 0
END) AS 'Unauthorized Individual(s)',
(CASE 
WHEN "Web Description" LIKE '%malware%' THEN '1'
ELSE 0
END) AS 'Malware',
(CASE
WHEN "Web Description" LIKE '%unsecured%' THEN '1'
ELSE 0
END) AS 'Unsecured'
FROM hack_df2;
''')

hack_df4 = mysql(query1)
hack_df4.info()
```

```python
# let's sum up all the columns; any that sum as 0 will later be classified as "N/A"

query2 = ('''
SELECT *, "Phishing" + "Ransomware" + "Cyberattack" + "Unauthorized Individual(s)" + "Malware" + "Unsecured" AS "Sum"
FROM hack_df4;
''')

hack_df5 = mysql(query2)
hack_df5.head()
```

```python
# let's label all the N/As

query3 = ('''
SELECT *, (CASE 
WHEN "Sum" = 0 THEN "N/A"
ELSE 0
END) AS "N/A"
FROM hack_df5;
''')

hack_df6 = mysql(query3)
hack_df6.head()
```

```python
# turn into .csv

hack_df6.to_csv('new - hack_types_with_NAs.csv')
```

```python
hack_df6['N/A'].value_counts()
```

```python
# all the #'s are objects, not numeric
hack_df6.info()
```

```python
# let's cast the #'s as numbers

query4 = ('''
SELECT "ID No.", "ID", "Investigation Status", "Name of Covered Entity", "State", "State Name", "Covered Entity Type", "Individuals Affected", "Breach Submission Date", "Type of Breach", "Location of Breached Information", "Business Associate Present", "Web Description", CAST("Phishing" AS INT) AS "Phishing", CAST("Ransomware" AS INT) AS "Ransomware", CAST("Cyberattack" AS INT) AS "Cyberattack", CAST("Unauthorized Individual(s)" AS INT) AS "Unauthorized Individual(s)", CAST("Malware" AS INT) AS "Malware", CAST("Unsecured" AS INT) AS "Unsecured", "Sum", "N/A"
FROM hack_df6;
''')

hack_df7 = mysql(query4)
hack_df7.head()
```

```python
hack_df7.info()
```

```python
# let's stick all hack types in one column

query5 = ('''
SELECT *, (CASE 
WHEN "Sum" > 1 THEN "Combination"
WHEN "Phishing" = 1 AND "Sum" = 1 THEN "Phishing Count"
WHEN "Ransomware" = 1 AND "Sum" = 1 THEN "Ransomware Count"
WHEN "Cyberattack" = 1 AND "Sum" = 1 THEN "Cyberattack Count"
WHEN "Unauthorized Individual(s)" = 1 AND "Sum" = 1 THEN "Unauthorized Individual(s) Count"
WHEN "Malware" = 1 AND "Sum" = 1 THEN "Malware Count"
WHEN "Unsecured" = 1 AND "Sum" = 1 THEN "Unsecured Count"
WHEN "Sum" = 0 THEN "N/A"
ELSE "Error"
END) AS "Hack Types"
FROM hack_df7;
''')

hack_df8 = mysql(query5)
hack_df8.head()
```

```python
hack_df8['Hack Types'].value_counts()
```

```python
# turn into .csv

hack_df8.to_csv('new - hack_keywords_one_col.csv')
```
